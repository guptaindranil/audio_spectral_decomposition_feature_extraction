```python
import librosa
import librosa.display
import scipy as sp
import IPython.display as ipd 
import matplotlib.pyplot as plt 
import numpy as np 
import os 
```

```python
debussy_file = "audio/voice.wav"
duke_file = "audio/duke.wav"
deepti_file = "audio/Deepti.wav"

debussy , sr = librosa.load(debussy_file)
rhcp, _ = librosa.load(rhcp_file)
duke, _ = librosa.load(deepti_file)
```

```python
FRAME_SIZE = 1024
HOP_SIZE = 512
```

```python
stft_debussy = librosa.stft(duke, n_fft = FRAME_SIZE, hop_length = HOP_SIZE)
stft_debussy.shape
```

```python
Y_debussy = np.abs(stft_debussy) ** 2
Y_debussy.shape
```

```python
def plot_spectogram (Y, sr, hop_length, y_axis = "linear"):
    plt.figure (figsize = (25 , 10))
    librosa.display.specshow(Y, sr = sr, hop_length = hop_length, x_axis = "time", y_axis= y_axis)
    plt.colorbar (format = "%+2.f")
    plt.savefig("log_spectogram.png")
```

```python
Y_log = librosa.power_to_db(Y_debussy)
plot_spectogram(Y_log, sr, HOP_SIZE)
```
